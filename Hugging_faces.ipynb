```python
# sentiment analysis
```

```python
from transformers import pipeline
classifer = pipeline('sentiment-analysis')
classifer('i hate this car')
```

```python
pipeline(task="sentiment-analysis", model="SamLowe/roberta-base-go_emotions")("i like  this car but i need some more features ")
```

```python
from transformers import pipeline
qa_pipeline = pipeline(
    "question-answering")
qa_pipeline(
    question="Where do I work?",
    context="My name is rajesh and i work as a fresher",
)
```
